```python
def Testpassword(password):
    if len(password) < 8 or len(password) > 14:
        return False
    digitnum = 0
    upletnum = 0
    spchanum = 0
    for i in range(len(password)):
        if password[i] in '0123456789':
            digitnum += 1
        if password[i] in 'ABCDEFGHIJKLMNOPQRSTUVWXYZ':
            upletnum += 1
        if password[i] in '!@#$%^&*()-_+=':
            spchanum += 1
    if digitnum < 2:
        return False
    if upletnum < 1:
        return False
    if spchanum < 1:
        return False
    return True

for i in range(3):
    p = raw_input()
    if Testpassword(p):
        print 'success'
        break
if not Testpassword(p):
    print 'error after 3 attempts'
```

```python
def expo(a, b):
    ans = a
    for i in range(b-1):
        ans *= a
    return ans

print expo(int(raw_input()), int(raw_input()))
```

```python
def min(numlist):
    ans = numlist[0]
    for i in range(len(numlist)):
        if numlist[i] < ans:
            ans = numlist[i]
    return ans

def max(numlist):
    ans = numlist[0]
    for i in range(len(numlist)):
        if numlist[i] > ans:
            ans = numlist[i]
    return ans
    
raw = raw_input()
nlist = raw.split()
for i in range(len(nlist)):
    nlist[i] = int(nlist[i])
print 'The smallest number is', min(nlist)
print 'The largest number is', max(nlist)
```
